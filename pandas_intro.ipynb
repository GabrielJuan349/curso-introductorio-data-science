## Ordenar un Dataframe

**df.sort_values(columna, ascending=boolean)**: *ascending*: True creciente, False decreciente **|** *columna*: Nombre de columna

**df.fillna(value=None)**: *value*: Valor por el cual lo quieres substituir 

**df.bfill()**

**df.ffill()**

**df.interpolate()**

**df.dropna()**

**df.groupby(columnas).get_group(valores)**

**df.groupby(columnas).agg(funciones)**

```python
print("Ordenar de forma ascendente por edad:")
df.sort_values('Age', ascending=True)
```

```python
print("Ordenar de forma descendente por edad:")
df.sort_values('Age', ascending=False)
```

```python
df_= pd.DataFrame({'x': [1,2,np.nan,4]})
df_
```

```python
print("Rellenar datos absdentes (NaN) con un 0:")
df_0 = df_.fillna(value=0)
df_0
```

```python
print("Rellenar datos NaN con el valor anterior:")
df_a= df_.bfill()
df_a
```

```python
print("Rellenar datos NaN con el valor posterior:")
df_p= df_.ffill()
df_p
```

```python
print("Hacer una interpolacion:")
df_i = df_.interpolate()
df_i
```

```python
print("Eliminar filas con NaN:")
df_NAN = df_.dropna()
df_NAN
```

```python
print("Agrupar por sexo y mostrar las mujeres:")
df_s_f = df.groupby('Sex').get_group('female')
df_s_f
```

```python
import numpy as np
print("Media de cada columna dividido por sexo:")
df_s_mean = df.groupby(["Sex"])["Pclass","Age","Price/Age","Fare","Fare/Mean","Survived","Siblings/Spouses Aboard","Parents/Children Aboard"].mean()
df_s_mean
```

```python
print("Extrae la media, el minimo y el maximo del precio del ticket segun la clase del pasajero:")
df_Pc_MMM = df.groupby("Pclass").agg({"Fare": ["min", "max","mean"]})
df_Pc_MMM
```
